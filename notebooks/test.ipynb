# An치lisis de datos de ventas

En este cuaderno haremos un an치lisis b치sico de ventas usando **Python** y **pandas**.

```python
import pandas as pd

# Crear un peque침o conjunto de datos
datos = {
    "Producto": ["A", "B", "C", "A", "B", "C"],
    "Ventas": [100, 150, 80, 120, 200, 90],
    "Mes": ["Enero", "Enero", "Enero", "Febrero", "Febrero", "Febrero"]
}

df = pd.DataFrame(datos)
df
```

```python
# Total de ventas por producto
ventas_por_producto = df.groupby("Producto")["Ventas"].sum()
ventas_por_producto
```
